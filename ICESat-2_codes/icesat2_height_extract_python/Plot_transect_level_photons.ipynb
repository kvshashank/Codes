```python
import h5py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
os.chdir('/home/shashank/Downloads/icesat2/icesat2_data/ATL03/sample_h5_cabo_rojo')
f = h5py.File("ATL03_20181028191317_04610101_004_01.h5", 'r')
f.visit(print)
seg_id = pd.DataFrame(f["gt1r/geolocation/segment_id"][:])
ref_ph_lat = pd.DataFrame(f["gt1r/geolocation/reference_photon_lat"][:])
ref_ph_lon = pd.DataFrame(f["gt1r/geolocation/reference_photon_lon"][:])

atl03 = pd.concat([seg_id, ref_ph_lat, ref_ph_lon], axis=1)
atl03.columns= ["seg_id", "ref_ph_lat", "ref_ph_lon"]

atl03 = atl03[(atl03["ref_ph_lat"] <= 18.08) & (atl03["ref_ph_lat"] >= 18.03) & (atl03["ref_ph_lon"] > -67.20) & 
              (atl03["ref_ph_lon"] < -66.78)].reset_index(drop=True)
print(atl03.shape)
```

```python
## Extract ATL08 data
os.chdir('/home/shashank/Downloads/icesat2/icesat2_data/ATL08/')
f = h5py.File("ATL08_20181028191317_04610101_004_01.h5", 'r')

classed_pc_flag = pd.DataFrame(f["gt1r/signal_photons/classed_pc_flag"][:])
classed_pc_indx = pd.DataFrame(f["gt1r/signal_photons/classed_pc_indx"][:])
d_flag = pd.DataFrame(f["gt1r/signal_photons/d_flag"][:])
delta_time = pd.DataFrame(f["gt1r/signal_photons/delta_time"][:])
ph_h = pd.DataFrame(f["gt1r/signal_photons/ph_h"][:])
ph_segment_id = pd.DataFrame(f["gt1r/signal_photons/ph_segment_id"][:])

atl08 = pd.concat([classed_pc_flag, classed_pc_indx, d_flag, delta_time, ph_h, ph_segment_id], axis=1)
atl08.columns = ["classed_pc_flag", "classed_pc_indx", "d_flag", "delta_time", "ph_h", "ph_segment_id"]

atl08 = atl08[atl08["ph_segment_id"].isin(atl03["seg_id"])]
atl08 = atl08.reset_index(drop=True)
atl08 = atl08[(atl08["ph_segment_id"] >= 100030) & (atl08["ph_segment_id"] <= 100063)].reset_index(drop=True)

print(atl08)
```

```python
plt.figure(figsize=(15,10))
toc = atl08[atl08["classed_pc_flag"] == 3].reset_index(drop=True)
canopy = atl08[atl08["classed_pc_flag"] == 2].reset_index(drop=True)
ground = atl08[atl08["classed_pc_flag"] == 1].reset_index(drop=True)


df_name = ground
new_df = pd.DataFrame([], columns = ["classed_pc_flag", "classed_pc_indx", "d_flag", "delta_time", 
                                     "ph_h", "ph_segment_id"])
for i in df_name["ph_segment_id"].unique():
    sub = df_name[df_name["ph_segment_id"] == i].reset_index(drop=True)
    sub = sub.sort_values("classed_pc_indx", ascending=True, ignore_index=True)
    new_df = new_df.append(sub, ignore_index=True)
plt.plot(np.arange(0, len(new_df["ph_h"])), new_df["ph_h"], "o", c="brown")

df_name=canopy
new_df = pd.DataFrame([], columns = ["classed_pc_flag", "classed_pc_indx", "d_flag", "delta_time", 
                                     "ph_h", "ph_segment_id"])
for i in df_name["ph_segment_id"].unique():
    sub = df_name[df_name["ph_segment_id"] == i].reset_index(drop=True)
    sub = sub.sort_values("classed_pc_indx", ascending=True, ignore_index=True)
    new_df = new_df.append(sub, ignore_index=True)
plt.plot(np.arange(0, len(new_df["ph_h"])), new_df["ph_h"], "o", c="green")


df_name=toc
new_df = pd.DataFrame([], columns = ["classed_pc_flag", "classed_pc_indx", "d_flag", "delta_time", 
                                     "ph_h", "ph_segment_id"])
for i in df_name["ph_segment_id"].unique():
    sub = df_name[df_name["ph_segment_id"] == i].reset_index(drop=True)
    sub = sub.sort_values("classed_pc_indx", ascending=True, ignore_index=True)
    new_df = new_df.append(sub, ignore_index=True)
plt.plot(np.arange(0, len(new_df["ph_h"])), new_df["ph_h"], "o", c="red")
plt.show()
```
